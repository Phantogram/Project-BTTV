```python
# Import dependencies
import os
import csv
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
# Read in CSV path

obesity_csvpath = os.path.join('data', 'obesity_state.csv')
obesity_csvpath

cancer_csvpath = os.path.join('data', 'cancer_death_state.csv')
cancer_csvpath


obesity_data = pd.read_csv(obesity_csvpath)
cancer_data = pd.read_csv(cancer_csvpath)

cancer_obesity = pd.merge(obesity_data, cancer_data, how="left", left_on='State', right_on="STATE")
```

```python
# Create a data frame
cancer_obesity.head()
```

```python
# Clean up the data
del cancer_obesity["URL"]
cancer_obesity
```

```python
# Clean up the data
del cancer_obesity["95% Confidence Interval"]
cancer_obesity.head()
```

```python
# Drop NaN
cancer_obesity.dropna(inplace=True)
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(cancer_obesity))
tick_locations = [value for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space,figsize is measured in inches
plt.figure(figsize=(14,6))

plt.bar(x_axis, cancer_obesity["RATE"], color='r', alpha=0.55, align="center")
plt.xticks(tick_locations, cancer_obesity["STATE"], rotation="vertical")

# Set x and y limits
plt.xlim(-0.75, len(x_axis))
plt.ylim(0, max(cancer_obesity["RATE"])+30)

# Set a Title and labels
plt.title("Cancer Death Rate by State")
plt.xlabel("State")
plt.ylabel("Cancer Death Rate")

# Save our graph and show the graph
plt.tight_layout()
plt.savefig("output_data/cancer_death.png")
plt.show()
```

```python
# Plot cancer death rate against obesity rate by state

x_values = cancer_obesity['RATE']
y_values = cancer_obesity['Prevalence']

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
print(f"r value: {rvalue}")
plt.plot(x_values,regress_values,"r-")

plt.scatter(x_values, y_values, s=80, edgecolor="black", alpha=0.75)
plt.xlabel("Cancer Death Rate by State")
plt.ylabel("Obesity Rate by State")
plt.title("Cancer Death Rate and Obesity Rate by State")

plt.savefig("output_data/obesity_cancer.png")
plt.show()
```
