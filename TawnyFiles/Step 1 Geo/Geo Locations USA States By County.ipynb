```python
# create a params dict that will be updated with new city each iteration
params = {"key": gkey}

# Loop through the cities_pd and run a lat/long search for each city
for index, row in df.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"

    county = row['County Name']
    state = row['STATE']

    # update address key value
    params['address'] = f"{county},{state}"

    # make request
    response = requests.get(base_url, params=params).json()
    
    try:
    
        df.loc[index, 'lat'] = response["results"][0]["geometry"]["location"]["lat"]
        df.loc[index, 'lng'] = response["results"][0]["geometry"]["location"]["lng"]
        
    except (KeyError, IndexError):
        df.loc[index, 'lat'] = 0
        df.loc[index, 'lng'] = 0
        print(f"Missing field/result... skipping {index}.")
```

```python
df["Ratings"] = df["AGE_ADJUSTED_RATE"].astype(float) / 100
df["Pop_Ratings"] = df["POPULATION"].astype(float) / 100000
df["Count_Ratings"] = df["COUNT"].astype(float) / 1000
```

```python
# Create map for cancer deaths in counties
locations = df[["lat", "lng"]]

# Fill NaN values and convert to float
rating = df["Ratings"].astype(float)

# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=rating, 
                                 dissipating=False, max_intensity=20,
                                 point_radius=1)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```
