```python
# Create code to answer each of the following questions.
# Hint: You will need multiple target URLs and multiple API requests.

# Dependencies
import requests
import json
import os
import csv
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
from scipy import stats
import gmaps

# Google API Key
from config import gkey
```

```python
# Specify path to read file
path = os.path.join("Resources", "USCS_OverviewMap.csv")

# Read in pandas
rain_df = pd.read_csv("Resources/avg_rain_state.csv")
deaths_df = pd.read_csv(path) 
deaths_df["State"] = deaths_df["Area"]
deaths_df["lat"] = ""
deaths_df["lng"] = ""

df = pd.merge(deaths_df, rain_df, on=["State"])
```

```python
# Build URL using the Google Maps API
base_url = "https://maps.googleapis.com/maps/api/geocode/json"

for index, row in df.iterrows():
        
    params = {"address": index, "key": gkey}
        
    response = requests.get(base_url, params=params).json()
    
    try:
    
        df.loc[index, 'lat'] = response["results"][0]["geometry"]["location"]["lat"]
        df.loc[index, 'lng'] = response["results"][0]["geometry"]["location"]["lng"]
        
    except (KeyError, IndexError):
        df.loc[index, 'lat'] = 0
        df.loc[index, 'lng'] = 0
        print(f"Missing field/result... skipping {index}.")
```

```python
df["Rating"] = df["AgeAdjustedRate"] / 100
```

```python
# df["Rating_pop"] = df["Population"] / 1000000
df.sort_values("lat")
```

```python
# Store latitude and longitude in locations
locations = df[["lat", "lng"]]

# Fill NaN values and convert to float
rating = df["Rating"].astype(float)

# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=rating, 
                                 dissipating=False, max_intensity=10,
                                 point_radius=1)


# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```
