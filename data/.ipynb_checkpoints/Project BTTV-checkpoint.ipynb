```python
import os
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
#study case files

#read to file
us_cancer_cases=os.path.join('../data','US_number_of_cases2.csv')
us_cancer_cases
```

```python
us_number_of_cancer_cases=pd.read_csv(us_cancer_cases)
us_number_of_cancer_cases.head()
```

```python
#us_number_of_cancer_cases["Area"]= us_number_of_cancer_cases["Area"].replace("'", "")
#us_number_of_cancer_cases
```

```python
#read alcohol consumption prevalence file
alcohol_consumption=os.path.join('../data','binge_drinkingPrev.csv')
alcohol_consumption
```

```python
binge_data=pd.read_csv(alcohol_consumption)
binge_data.head()
```

```python
merged_df=pd.merge(us_number_of_cancer_cases,binge_data,how='left',left_on='Area',right_on='State')
merged_df
```

```python
# Calculate the correlation coefficient between new cancer cases and prevalence of binge drinking

Percentage = merged_df.iloc[:,9]
AgeAdjustedRate = merged_df.iloc[:,5]
correlation = st.pearsonr(Percentage,AgeAdjustedRate)
#plot a scatter plot
plt.scatter(merged_df["Percentage"],merged_df["AgeAdjustedRate"],facecolors="red", edgecolors="black",alpha=0.75)
# Add labels to the x and y axes
plt.title("Comparison between Cancer Cases and Prevalence of Binge Drinking in US")
plt.xlabel("Prevalence of Binge drinking in Percentage")
plt.ylabel("AgeAdjustedRate")
print(f"The correlation between cancer cases and prevalence of binge drinking percentage {round(correlation[0],2)}")
plt.show()
```

```python
#read cancer death cases file
cancer_deaths=os.path.join('../data','cancer_deaths_2017.csv')
cancer_deaths
```

```python
cancer_death_data=pd.read_csv(cancer_deaths)
cancer_death_data.head()
```

```python
cancer_death_data=pd.read_csv(cancer_deaths)
del cancer_death_data['URL']
cancer_death_data.head()
```

```python
#merge cancer death with alcohol binge drinking prevalence
merge_death_cases=pd.merge(cancer_death_data,binge_data,how='left',left_on='STATE',right_on='State')
merge_death_cases.head()
```

```python
# Calculate the correlation coefficient between cancer deaths and prevalence of binge drinking
Percentage = merge_death_cases.iloc[:,5]
Rate = merge_death_cases.iloc[:,2]
correlation = st.pearsonr(Percentage,Rate)
# scatter plot
plt.scatter(merge_death_cases["Percentage"],merge_death_cases["RATE"],facecolors="green", edgecolors="black",alpha=0.75)
# Add labels to the x and y axes
plt.title("Cancer deaths and Prevalence of Binge Drinking by State")
plt.xlabel("Prevalence of Binge drinking in Percentage")
plt.ylabel("Cancer Death Rate")
print(f"The correlation between prevalence of binge drinking percentage and cancer deaths is {round(correlation[0],2)}")
plt.savefig('Output_folder/deathsbinge_drinking.png')
plt.show()
```

```python
#read wine_data file
wine_data=os.path.join('../data','wine_data_states.csv')
wine_data
```

```python
wine_consumption=pd.read_csv(wine_data)
wine_consumption.head()
```

```python
#merge wine data and cancer death data to single data frame
merged_wine_data=pd.merge(cancer_death_data,wine_consumption,how='left',left_on='STATE',right_on='State')
merged_wine_data.head()
```

```python
# Calculate the correlation coefficient between cancer deaths and wine consumption vinepair
Gallons_of_Ethanol_Per_Capita = merged_wine_data.iloc[:,6]
Rate = merged_wine_data.iloc[:,2]
correlation = st.pearsonr(Gallons_of_Ethanol_Per_Capita,Rate)
# scatter plot
plt.scatter(merged_wine_data["Gallons of Ethanol Per Capita*"],merged_wine_data["RATE"],facecolors="blue", edgecolors="black",alpha=0.75)
# Add labels to the x and y axes
plt.title("Comparison between cancer deaths and wine consumption in US by State")
plt.xlabel("Gallons of Ethanol Per Capita*")
plt.ylabel("Rate")
print(f"The correlation between wine consumption and cancer deaths is {round(correlation[0],2)}")
plt.show()
```

```python
#read beer data in file
beer_data=os.path.join('../data','beer_data.csv')
beer_data
```

```python
#read in to beer data
beer_consumption=pd.read_csv(beer_data)
beer_consumption.head(5)
```

```python
#merge beer data with cancer deaths to single dataframe
merged_beer_data=pd.merge(cancer_death_data,beer_consumption,how='left',left_on='STATE',right_on='State')
merged_beer_data.head()
```

```python
# Calculate the correlation coefficient between cancer deaths and beer consumption vinepair
Gallons_of_Ethanol_Per_Capita = merged_beer_data.iloc[:,6]
Rate = merged_beer_data.iloc[:,2]
correlation = st.pearsonr(Gallons_of_Ethanol_Per_Capita,Rate)
# scatter plot
plt.scatter(merged_beer_data["Gallons of Ethanol Per Capita*"],merged_beer_data["RATE"],facecolors="black", edgecolors="black",alpha=0.75)
# Add labels to the x and y axes
plt.title("Comparison between cancer deaths and beer consumption in US by State")
plt.xlabel("Gallons of Ethanol Per Capita*")
plt.ylabel("Rate")
print(f"The correlation between both factors is {round(correlation[0],2)}")
plt.show()
```
